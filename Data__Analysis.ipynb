```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load your preprocessed data
df = pd.read_csv('/content/drive/MyDrive/BugSum-master/excel sheet/Preprocessed_Data.csv')
```

```python
# Word Cloud for Titles
title_wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(' '.join(df['Title']))
plt.figure(figsize=(10, 5))
plt.imshow(title_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Titles')
plt.show()
```

```python
# Word Cloud for ProcessedText using map in one line
text_wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate_from_text(' '.join(df['ProcessedText'].map(lambda x: str(x) if pd.notna(x) else '')))

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(text_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for ProcessedText')
plt.show()
```

```python
# Summary Length Analysis
df['Title_Length'] = df['Title'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)
df['ProcessedText_Length'] = df['ProcessedText'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)
```

```python
# Histograms for Lengths
plt.figure(figsize=(12, 6))
sns.histplot(df['Title_Length'], bins=30, kde=True, color='blue', label='Title Length')
sns.histplot(df['ProcessedText_Length'], bins=30, kde=True, color='orange', label='ProcessedText Length')
plt.title('Distribution of Title and ProcessedText Lengths')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Pair Plot for Lengths and Bug Numbers
sns.pairplot(df[['Title_Length', 'ProcessedText_Length', 'BugNum']])
plt.suptitle('Pair Plot for Lengths and Bug Numbers', y=1.02)
plt.show()
```
